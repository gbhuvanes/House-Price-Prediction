```python
data_refine = data.drop('ocean_proximity', axis = 1)
data_refine
```

```python
data_refine = data_refine.dropna(axis = 0)
data_refine
```

```python
dt = data_refine.groupby("housing_median_age", as_index=False) 

groups = dict(list(dt))

df=pd.DataFrame.from_dict(groups,orient='index').transpose()

pd.DataFrame(dt.size().reset_index(name = "Group_Count"))
gp = dt.median()

gp1 = dt.loc("housing_median_age") < 25)
print(dt.index)
```

```python
# X = data_refine.drop('median_house_value', axis = 1)
X = data_refine[['total_rooms','housing_median_age']]
Y = data_refine['median_house_value']
print(Y)
```

```python
X.info()
```

```python
scatter = plt.scatter(X['housing_median_age'],X['total_bedrooms'],X['median_income'], Y)
plt.xlabel('X')
plt.ylabel('House Price')
classes = ['housing_median_age', 'total_bedrooms', 'median_income']
plt.legend(handles=scatter.legend_elements()[0], labels=classes)

# data_refine = data_refine.dropna(axis = 0)
# data_refine.info()
```

```python
X.head()
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')
X['total_bedrooms'].plot.box(color=color)
```

```python
LR = slm.LinearRegression()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
LR.fit(X_train, Y_train)
```

```python
X_test
```

```python
predict = LR.predict(X_test)
predict
```

```python
print('Predicted Value :',predict[0])
print('Actual Value :',Y_test.values[0])
```

```python
LR.score (X_test, Y_test)
```

```python
error = predict[0]- Y_test.values[0]
error
```

```python
gr = pd.DataFrame({'Predicted':predict,'Actual':Y_test})
gr = gr.reset_index()
gr = gr.drop(['index'],axis=1)
plt.plot(gr[:1000])
plt.legend(['Actual','Predicted'])
#gr.plot.bar();
```
